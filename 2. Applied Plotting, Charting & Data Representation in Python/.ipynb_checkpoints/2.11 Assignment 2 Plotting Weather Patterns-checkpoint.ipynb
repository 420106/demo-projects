```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.dates import DateFormatter, MonthLocator
```

```python
df = pd.read_csv('data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month-Day'] = df['Date'].dt.strftime('%m-%d')

# Convert to degrees C
df['Data_Value'] = df['Data_Value'] / 10

# Remove leap year days
df = df[df['Month-Day'] != '02-29']

# Split 2015 data
df_2015 = df[df['Year'] == 2015]
df = df[df['Year'] < 2015]

# Record high and low from 2005 to 2014
high = df[df['Element'] == 'TMAX'].groupby('Month-Day').max()['Data_Value']
low = df[df['Element'] == 'TMIN'].groupby('Month-Day').min()['Data_Value']

# Record high and low in 2015
high_2015 = df_2015[df_2015['Element'] == 'TMAX'].groupby('Month-Day').max()['Data_Value']
low_2015 = df_2015[df_2015['Element'] == 'TMIN'].groupby('Month-Day').min()['Data_Value']

# Concatenate all the 4 series
df = pd.concat([high, low, high_2015, low_2015], axis = 1)
df.columns = ['High', 'Low', 'High_2015', 'Low_2015']

# Identify the broken record
df['High_Broken'] = df[df['High_2015'] > df['High']]['High_2015']
df['Low_Broken'] = df[df['Low_2015'] < df['Low']]['Low_2015']

df = df.reset_index()

df.head()
```

```python
# %matplotlib notebook

plt.figure(figsize = (18, 7))

x_range = np.arange('2015-01-01', '2016-01-01', dtype = 'datetime64')

# Record high
plt.plot(x_range, df['High'], color = '#F9A7B0')
''' Lava Red: #E42217
    Flamingo Pink #F9A7B0
    Pink Bubble Gum #FFDFDD '''

# Record low
plt.plot(x_range, df['Low'], color = '#A0CFEC')
''' Blue Ribbon: #306EFF
    Jeans Blue: #A0CFEC
    Robin Egg Blue: #BDEDFF '''

# Overlay 2015 data where records are broken
plt.scatter(x_range, df['High_Broken'], color = '#E42217')
plt.scatter(x_range, df['Low_Broken'], color = '#306EFF')

# Fill it up
ax = plt.gca()
ax.fill_between(x_range, df['High'], df['Low'], facecolor = '#FFEBCD', alpha = 0.6)
''' Platinum: #E5E4E2
    Cornsilk: #FFF8DC
    Blached Almond: #FFEBCD'''

# Clean it up
for spine in plt.gca().spines.values():
    spine.set_visible(False)

plt.tick_params(top='off', bottom='off', left='off', right='off')

# Label it up
plt.legend(['Historical High', 'Historical Low', '2015 High', '2015 Low'])
plt.title('Weather Patterns: Historical Record vs. 2015 Breaking Record' + '\n' + '(Ann Arbor, Michigan, United States)')
plt.ylabel('Temperature (C)')
plt.axes().yaxis.grid(linestyle = '--', alpha = 0.8)

# Convert x-axis value to month
ax.xaxis.set_major_locator(MonthLocator())
ax.xaxis.set_major_formatter(DateFormatter('%b'))


plt.show()
```
