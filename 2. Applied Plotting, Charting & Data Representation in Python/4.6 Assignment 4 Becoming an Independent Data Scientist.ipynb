```python
import os
import numpy as np
import pandas as pd

path = 'Assignment 4 - data/'
os.listdir(path)
```

### Unaffiliated Population vs All Population 

```python
''' Population datasets are observationally formatted in the exact same way. All the four datasets are in the same shape, and 
    share the same ['iso_code'] and [' name'] if sorted '''

pd.read_csv(path + 'WorldwideAllPopulation2020_.csv', skiprows = 7, skipfooter = 3, engine = 'python').head()
```

```python
''' To load and merge all population datasets on ISO code and region name, and rename the columns accordingly '''

dfs = [] 
for filename in ['WorldwideUnaffiliatedPopulation2010_.csv', 'WorldwideAllPopulation2010_.csv', 
                 'WorldwideUnaffiliatedPopulation2020_.csv', 'WorldwideAllPopulation2020_.csv']:
    dfs.append(pd.read_csv(path + filename,
                           skiprows = 7, skipfooter = 3, engine = 'python') # engine='python' to avoid ParserWarning
               .set_index(['iso_code', ' name'])
              )
df = pd.concat(dfs, axis = 1).reset_index()
df.columns = ['ISO code', 'Region', 'Unaffiliated 2010', 'All 2010', 'Unaffiliated 2020', 'All 2020']

print(df.shape)
df.head()
```

```python
''' Seven NaN ISO code while six of them are continent level regions to be removed. An intersting finding is that Namibia is 
    with ISO code "NA" which is somehow converted into a NaN value '''

df['Region'][df['ISO code'].isnull()].tolist()
```

```python
''' To clean the data with the following steps:

    1. Remove continent level regions with NaN value while keep Namibia;
    2. Remove all observations with < 10000 unaffiliated population as these data would be less accurate
    3. Drop all ['ISO_code'] columns as there is no further use;
    4. Trim space and double quotes throughout the entire dataframe;
    5. Convert population data from string to integer '''

nan_rows = df['Region'][df['ISO code'].isnull()].tolist()
nan_rows.remove(' "Namibia"')
df = (df[(~df['Region'].isin(nan_rows)) & 
         (df['Unaffiliated 2010'] != ' "<10000"') &
         (df['Unaffiliated 2020'] != ' "<10000"')]
      .drop('ISO code', axis = 1)
      .apply(lambda x: x.str.strip(' ').str.strip('"'))
)
df[['Unaffiliated 2010', 'All 2010', 'Unaffiliated 2020', 'All 2020']] = (
 df[['Unaffiliated 2010', 'All 2010', 'Unaffiliated 2020', 'All 2020']].astype(int)
)

print(df.shape)
df.head()
```

```python
''' To estimate 2016 unaffliliated ratio with the following two steps:
    
    1. To estimate 2016 unaffliliated and whole population using the sum of 2010 and 2020 data multiplied with 0.6
    2. To work out 2016 unaffiliated ratio, round to 3 decimal places '''

df['Unaffiliated 2016'] = 0.6 * (df['Unaffiliated 2010'] + df['Unaffiliated 2020'])
df['All 2016'] = 0.6 * (df['All 2010'] + df['All 2020'])
df['Unaffiliated Ratio 2016'] = (df['Unaffiliated 2016'] / df['All 2016']).round(3)

print(df.shape)
df.head()
```

```python
''' ['Region'] and ['Unaffiliated Ratio 2016'] are the only two columns needed so far for the later analysis '''

df = df[['Region', 'Unaffiliated Ratio 2016']]

print(df.shape)
df.head()
```

### SHI vs GRI

```python
pd.read_csv(path + 'SocialHostilitiesIndexScoreIn2016_.csv', skiprows = 3, skipfooter = 3, engine = 'python').head()
```

```python
''' To left join SHI and GRI data to the main dataframe '''

df = (pd.merge(df,
              pd.read_csv(path + 'SocialHostilitiesIndexScoreIn2016_.csv',
                          skiprows = 3, skipfooter = 3, engine = 'python')
              .rename(columns = {'name': 'Region'}), 
              how = 'left', on = 'Region'
             )
      .drop_duplicates() # Each entry appears twice
     )

print(df.shape)
df.head()
```

```python
''' To further clean the dataframe with the following steps:

    1. Remove observations with no values of SHI and GRI
    2. Rename the columns with shorter titles
    3. Sort the dataset by the unaffilidated ratio descending '''

df = (df.dropna()
      .rename(columns = {'Unaffiliated Ratio 2016': 'Ratio',' Social': 'SHI', ' Government': 'GRI'})
      .sort_values('Ratio', ascending = False)
      .set_index('Region')
      .reset_index()
     )

print(df.shape)
df.head()
```

### Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Patch
```

```python
''' To inspect the correlation of unaffiliated ratio against SHI and GRI '''

# Scatter plot
plt.figure(figsize = (18, 6))
plt.scatter(x = df['Ratio'], y = df['SHI'], 
            marker = '_', color = 'red', alpha = 0.6)
plt.scatter(x = df['Ratio'], y = df['GRI'], 
            marker = '|', color = 'blue', alpha = 0.6)
plt.show()

# Correlation
corr_shi = df.corr().loc['Ratio', 'SHI']
corr_gri = df.corr().loc['Ratio', 'GRI']
print('Ratio/SHI: ', corr_shi)
print('Ratio/GRI: ', corr_gri)
```

```python
''' To visualize SHI, GRI, and unaffiliated ratio in a single bubble chart '''

plt.figure(figsize = (12, 12))

# Top 10 country with high unaffilidated ratio
plt.scatter(x = df['SHI'][:15], y = df['GRI'][:15], s = df['Ratio'][:15] * 4500, 
            color = 'yellow', edgecolors = 'red', linewidths = 3, alpha = 0.65
           )

# All the rest countries
plt.scatter(x = df['SHI'][15:], y = df['GRI'][15:], s = df['Ratio'][15:] * 4500, 
            color = 'blue', alpha = 0.2
           )

# Label
plt.title('Irreligion Ratio vs. SHI & GRI' + '\n' + 
          '(149 regions worldwide, 2016)' + '\n',
          fontsize = 14
         )
plt.xlabel('Social Hostilities Index (SHI)', fontsize = 13)
plt.xticks(fontsize = 13)
plt.ylabel('Government Restrictions Index (GRI)', fontsize = 13)
plt.yticks(fontsize = 13)

# Legend
lgd = plt.legend(loc = 'lower right', 
                 labels = ['Top 15 regions (top 5 annotated)', 'The rest regions'],
                 fontsize = 12
                )
for handle in lgd.legendHandles:
    handle._sizes = [150] # Resize legend marker for consistency

# Annotation
bbox_props = dict(boxstyle = 'square', facecolor = 'black', alpha = 0.9)
arrow_props = dict(arrowstyle = '-', linewidth = 1, facecolor = 'black')
for i in range(5):
    plt.annotate(xy = (df['SHI'][i] + df['Ratio'][i] / 2.5, df['GRI'][i]),
                 xytext = (df['SHI'][i] + 0.5, df['GRI'][i] + 0.2),
                 s = df['Region'][i],
                 fontweight = 'semibold', color = 'white',
                 bbox = bbox_props, arrowprops = arrow_props
                )

# Additional List
text = 'Top 15 regions:' + '\n\n'
for i in range(15):
    text = text + '{}:  {}%\n'.format(df['Region'][i], 
                                      round(df['Ratio'][i] * 100,2)
                                     )
text = text + '\n' + 'Note: ratio represented by bubble size' + '\n'
bbox_props = dict(facecolor = 'lightskyblue', alpha = 0.15)
plt.text(x = 10.5, y = 3, s = text, fontsize = 13, linespacing = 2, bbox = bbox_props)

plt.show()
```
